```python
import matplotlib.pyplot as plt

student1 = {'chinese':78, 
            'english':80,
            'math':92,
            'history':75,
            'discover':85}

student2 = {'chinese':98, 
            'english':100,
            'math':98,
            'history':97.5,
            'discover':100}

subjects = list(student1.keys())
score1 = list(student1.values())

figure = plt.figure(figsize=(8,5))
axes = figure.add_subplot(1,1,1)
axes.plot(subjects,score1,'ro--',label='Sudent1')

score2 = list(student2.values())
axes.plot(subjects,score2,'bo-',label='Student2')

axes.set_title('Table Of Students')
axes.set_xlabel('Subject')
axes.set_ylabel('Scores')
axes.grid(axis='y')
axes.legend()
xlocation=axes.get_xticks()

print(xlocation)
for i in xlocation:
    axes.annotate(str(score1[i]),xy=(i,score1[i]+0.5))
    axes.annotate(str(score2[i]),xy=(i,score2[i]+0.5))

plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties
font=FontProperties(fname='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc',size=8)

student1 = {'':78, 
            '':80,
            '':92,
            '':75,
            '':85}

student2 = {'':98, 
            '':100,
            '':98,
            '':97.5,
            '':100}

subjects = list(student1.keys())
score1 = list(student1.values())

figure = plt.figure(figsize=(8,5))
axes = figure.add_subplot(1,1,1)
axes.plot(subjects,score1,'ro--',label='1')

score2 = list(student2.values())
axes.plot(subjects,score2,'bo-',label='2')

axes.set_title('',fontproperties=font)
axes.set_xlabel('',fontproperties=font)
axes.set_ylabel('',fontproperties=font)
axes.set_xticklabels(subjects,fontproperties=font)
axes.grid(axis='y')
axes.legend(prop=font)
xlocation=axes.get_xticks()

print(xlocation)
for i in xlocation:
    axes.annotate(str(score1[i]),xy=(i,score1[i]+0.5))
    axes.annotate(str(score2[i]),xy=(i,score2[i]+0.5))

plt.show()
```
