```python
import random
import numpy as np
import pandas as pd

with open("names.txt",encoding="utf-8") as file:
    content = file.read()

names_list = content.split()
names = random.choices(names_list,k=50)

with open("names.txt",encoding="utf-8") as file:
    content = file.read()
    
scores = np.random.randint(50,101,size=(50,5))
student_df = pd.DataFrame(scores,
             columns=['國文','英文','數學','地理','歷史'],
             index=names)
student_df
```

```python
sum=student_df.sum(axis=1)
mean=student_df.mean(axis=1)
```

```python
student_df['總分']=sum
student_df['平均']=mean
student_df
```

```python
sorted=student_df.sort_values(by='總分',ascending=False)
sorted['排名']=sorted["總分"].rank(ascending=False,method='min')
```

```python
sorted
```

```python
def fun1(s):
    s1=s[['國文','英文','數學','地理','歷史']]
    less60=s1[s1<60].count()
    return less60
series=sorted.apply(fun1,axis=1)
sorted['不及格的科目數']=series
```

```python
def fun2(s):
    s1=s[['國文','英文','數學','地理','歷史']]
    less60=s1[s1<60]
    return less60.index.values
sorted['不及格科目']=sorted.apply(fun2,axis=1)
```
